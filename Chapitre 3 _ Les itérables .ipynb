#### **Le slicing de séquences en Python**

- Cela consiste à extraire une sous-séquence à partir d'une séquence.
- Le slicing fonctionne de manière similaire aux intervalles mathématiques : `[début:fin[`
- La borne de fin ne fait pas partie de l'intervalle sélectionné.
- La syntaxe générale est `L[i:j:k]`, où :
    - `i` = indice de début
    - `j` = indice de fin, élément non sélectionné
    - `k` = le "pas" ou intervalle (s'il est omis alors il vaut `1`)
- La sous-liste sera donc composée de tous les éléments de l’indice `i` jusqu’à l’indice `j-1`, par pas de `k`.
- La sous-liste est un nouvel objet.

Dans le sens normal (le pas `k` est positif)

- Si `i` est omis alors il vaut `0`
- Si `j` est omis alors il vaut `len(L)`

Dans le sens inverse (le pas `k` est négatif)

- Si `i` est omis alors il vaut `-1`
- Si `j` est omis alors il vaut `-len(L)-1`

Illustrons ça en créant une liste à partir d'une chaîne de caractères.

La fonction [`split()`](https://docs.python.org/3/library/stdtypes.html#str.split) découpe une chaîne de caractères en morceaux, par défaut en 'mots'.

````python
ZenOfPython = '''Flat is better than nested.
Sparse is better than dense.
Readability counts.''' 
L = ZenOfPython.split()
print(L)
````

Pour commencer, on extrait de la liste ``L`` un nouvel objet liste qui contient tous les éléments de ``L`` <=> copie de liste

````python
# Les trois écritures sont équivalentes
print(L[0:len(L):1]) # Cette notation est inutilement lourde car :
print(L[::])         # i = 0, j=len(L) et k=1 donc i, j et k peuvent être omis
print(L[:])          # on peut même ommettre le 2ème ":"
````

On extrait une sous-liste qui ne contient que les 4 premiers éléments :

````python
# Les trois écritures sont équivalentes
print(L[0:5:1])  # Notation complète
print(L[:5:1])  # Le premier indice vaut i=0, donc on peut l'omettre
print(L[:5])  # Le pas de slicing vaut 1, donc on peut l'omettre, ainsi que le ":"
````

Les indices peuvent être négatifs, ce qui permet de traiter les derniers éléments :

````python
# on veut exclure les derniers éléments :
print(L[0:-2:1])  # Notation complète
print(L[:-2:1])  # Le premier indice vaut i=0, donc on peut l'omettre
print(L[:-2])  # Le pas de slicing vaut 1, donc on peut l'omettre
````

On ne garde que les deux derniers éléments

````python
print(L[-2:])
````

Le slicing peut être utilisé sur des chaînes de caractères.

````python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
print(alphabet[:4], alphabet[-4:], alphabet[0:12:3])
````

#### **Types muables et types immuables**

Avant d'aller plus loin dans la revue des types, il est important de comprendre le mécanisme d'affectation en fonction du caractère muable ou immuable de l'objet.

#### Cas d'un objet muable

Deux noms de variables différents peuvent référencer le même objet.
Si cet objet est muable, les modifications faites par l'intermédiaire d'une des variables sont visibles par toutes les autres.

````python
a = ['Cerise', 'Citron']  # on initialise la variable a
b = a                # b référence le même objet que a
````

`a` et `b` possèdent la même référence :

````python
print(id(a))
print(id(b))
````

`a` et `b` contiennent la même donnée :

````python
print(a)
print(b)
````

Si on modifie la donnée de `a`, la donnée de `b` est aussi modifiée !

````python
a.append('Amande')
print(a)
print(b)
````

#### Cas d'un objet immuable


````python
t = 'banane', 'orange'
u = t
print(id(t))
print(id(u))
````

`t` et `u` sont deux variables qui référencent le même objet `tuple` donc leur donnée ne peut être modifiée.
Tout ce qu'on peut faire, c'est affecter une nouvelle valeur :

````python
t = 'Amande', 'citron'
````

Dans ce cas, `t` référence un nouvel objet alors que `u` référence toujours l'objet initial :

````python
print(id(t))
print(id(u))
````

Et bien sûr leurs données sont différentes :

````python
print(t)
print(u)
````

#### Un peu plus loin...

Bien que sa séquence soit immuable, si un `tuple` est constitué d'éléments muables, alors ces éléments-là peuvent être modifiés.

Illustration avec un `tuple` dont un des éléments est une `list` :

````python
T = ('a', ['b', 'c']) # le deuxième élément est une liste donc il est mutable
print('T =', T)
L = T[1]
print('L =', L)
L[0] = 'e'
print('L =', L)
print('T =', T)
````

````python
# Ici on fait exactement la même chose...
T = ('a', ['b', 'c'])
print('T =', T)
T[1][0] = 'z'
print('T après =', T)
````

Exemple d’exécution :  L=[1,2,5,8,6,2,5,9,1,8,8] 


````python
def supprimer_redoublant(L):
    """ supprimer les redoublant dans  une  Liste 
    """
    i=0
    while i<len(L) :  # parcourt de la  liste L 
        j=i+1    # on commence la  recherche des doublant de L[i]
        while j<len(L):
            if L[i]==L[j]:  #suppression du redoublaant rouvé à l'indice j
                L.pop(j) #la taille se reduite automatiquement donc il ne faut pas d'incrémenter l'indice j
            else:
                j+=1 #incrémentation de l'indice de j si l[j]et différent que L[i]
        i+=1
    return L

L=[1,2,5,8,6,2,5,9,1,8,8]
print(supprimer_redoublant(L))
````

````python
def inserer(val,L):
    """
    inserer(val,L) permet d'insérer la valeur val dans la liste triée L
    """
    L.append(val)#ajouter une valeur (n'importe quelle valeur) pour agrandirla liste
    n=len(L)#récupérer la taille de la liste L
    i=n-2 #initiation de i par n-2 (l'indice de l'avant dernier élément)
    
    #la boucle while cherche l'emplacement de la valeur val et on décale au même temps
    while i>=0 and L[i]>val: 
        L[i+1]=L[i]  # décalage
        i-=1         # aller vers l'élément précédent 
        """on commence par l'avant-dernier élément de la liste et on sort de la boucle au premier élément trouvé qui est inférieur à val, 
        l'indice de cette élément est i
        """
    L[i+1]=val
    return L

L=[ 2,4,8,10,15,19,20,25]
inserer(7,L)
print(L)
````

Pour tester la fonction on va créer la procédure **affichage_canonique( P )** qui affiche sous une forme canonique (en débutant des exposants les plus grands vers les plus petits)

Un polynôme peut être représenté par une liste dont les éléments sont les coefficients du polynôme et les indices sont les exposants.
Le polynôme $P(x) = x³ + 2x² – x+5$   peut présenté par la liste P=[ 1 , 2 , -1 , 5 ]

Exemple: Calcul du produit de $P(x)$ et $Q(x)$ avec

$P(x) = x³ + 2x² – x+5$
et

$Q(x) = 3x²  – 2$


````python
def affiche_canonique(P):
    l=len(P)
    pr=1
    s=c=p=x=""
    #s pour le signe de la coefficient
    #c pour la valeur de la coefficient
    #p pour ^degré
    #x pour le symbole de x
    for i in range(l-1,-1,-1):
         
        if P[i]!=0:
            if pr==1:
                if P[i]>0:s=""
                if P[i]<0:s=" - " else: if P[i]>0:s=" + "
                if P[i]<0:s=" - " pr=0 c=str(abs(P[i])) if abs(P[i])==1 and i!=0:c="" if i==0:x="" else: x="x" if i>1:p="^"+str(i)
            else:p=""
            print(s,c,x,p,sep="",end="");
    print()
````

````python
def produit_polynome(P,Q):
    d_p=len(P)-1 # le degré de P
    d_q=len(Q)-1 # le degré de Q
     
    PQ=[0 for i in range(d_p + d_q+1)] # création de la polynôme PQ=PxQ de degré d_pq = d_p + d_q
    for i in range(d_p+d_q+1):
        coeff_PQ_deg_i=0 #coeff_deg_i pour calculer le coefficient de la polynôme PQ de degré i
        for j in range(i+1):
            if j<=d_p : coeff_P_deg_j=P[j]
            else:coeff_P_deg_j=0
            if (i-j)<=d_q : coeff_Q_deg_j=Q[i-j]
            else: coeff_Q_deg_j=0
            coeff_PQ_deg_i += coeff_Q_deg_j*coeff_P_deg_j
        PQ[i]=coeff_PQ_deg_i
    return PQ
````
